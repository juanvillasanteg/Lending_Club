## 3.1. Numerical features

We develop the same procedure for each of the numerical features.
First, we see the descriptions and the histogram of frequencies to analyze the data. Then, we impute the missing values through the median, random imputation or othe method and finally we check if there are outliers and establish criteria to eliminate them.

```python
def fill_median(data, column):
    data[column].fillna(data[column].median(), inplace = True)
```

```python
def histogram(variable, bins=50):
    fig = sns.distplot(cleandf[f"{variable}"] ,hist=True, bins=bins, kde=False)
    return fig
```

```python
def distribution(variable):
    fig = cleandf.groupby(f'{variable}')['target'].value_counts(normalize=True).loc[:,1].plot(kind='bar')
    return fig
```

```python
def nulls(variable):
    return sum(cleandf[f'{variable}'].isnull())
```

```python
#specify the metric column names to be modelled
numeric = cleandf.select_dtypes(include = [np.number])
to_model_columns = numeric.columns
clf=IsolationForest(n_estimators=100, max_samples='auto', contamination=float(.04), \
                        max_features=1.0, bootstrap=False, n_jobs=5, random_state=42, verbose=0)
clf.fit(cleandf[to_model_columns])
pred = clf.predict(cleandf[to_model_columns])
cleandf['anomaly']=pred
outliers=cleandf.loc[cleandf['anomaly']==-1]
outlier_index=list(outliers.index)
#print(outlier_index)
#Find the number of anomalies and normal points here points classified -1 are anomalous
print(cleandf['anomaly'].value_counts())
```

```python
cleandf = cleandf[cleandf['anomaly'] != -1]
```
