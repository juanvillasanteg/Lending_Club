Our target column is called `loan_status`, so lets check which categories include this feature and the frecuency each category.

```python
df["loan_status"].value_counts()
```

After analyzing the description of `loan_status` we decided to remove "Current", data of the individuals who are up to date with payments on the repayment of the loan becouse we don't know if they are good or bad payers. For the same reason we also decided to remove the category "In Grace Period".

```python
df = df[df["loan_status"] != "Current"]
```

```python
df = df[df["loan_status"] != "In Grace Period"]
```

```python
df.shape
```

Furthermore we are going to create a target based on `loan status`. We will encode `loan_status` as our target with the clients who have completed the loan as 0 (good clients) and the rest will be considered as 1 (bad clients). Then we check the proportion of good and bad clients in the dataset and delete the original feature.

```python
df["target"] = np.where((df['loan_status'] == 'Does not meet the credit policy. Status:Fully Paid') |
                        (df['loan_status'] == 'Fully Paid'), 0, 1)
```

```python
sns.barplot(x="target", y="target", data=df, estimator=lambda x: len(x) / len(df) * 100);
```

```python
print("Percetage of people that have paid: ", round(sum(df["target"]/len(df["target"])),2))
```

```python
print("Percetage of people that haven't paid: ", round((1-sum(df["target"]/len(df["target"]))),2))
```

```python
df['target'].value_counts(normalize=True, dropna=False)
```

```python
df.drop("loan_status", axis=1, inplace=True)
```
