# 0. Problem statment

## 0.1 About the Dataset

These files contain loan data in the 2007-2019 period, including the current loan status and the latest payment information.

Features include number of financial inquiries, zip code addresses and status. The file is a matrix of 2.260.668 observations and 145 features. A data dictionary is provided in a separate file.

## 0.2. Goal and Questions

Our goal is to build a machine learning algorithm for the prediction of loan granting in the current features in the database. To achieve this, we will perform an analysis of the database trying to answer the following questions:

1) What type of features do we have in the dataset?

2) Are there any features that filter information or aren't relevant?

3) What degree of relationship do the variables have? Are there missing values? How do we impute them?

4) What variables are more important in prediction models?

5) What prediction model shows better results?

## 0.3. Results and Conclusions

Answering the previous questions we have detected that there are three kinds of data in the dataset, strings, booleans and numerics. All of all features included in the dataset (145) we have keeped only a few of them which we consider they have the most relevant information for for the goal we have set.

Many features included in the dataset filtered information about our objective to predict because they were collected at a time after the credit has been granted. 

Also a large number of features were incomplete, having missing vaules for more than 60% of the records.

After performing multiple machine learning models, we represent the combination of ROC curves obtained by the models through a plot.
